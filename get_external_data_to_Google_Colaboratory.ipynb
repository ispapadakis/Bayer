# Get Directly from *URL*

```
# Example UCI Data

url_data = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
url_metadata = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.names"
```

```
import requests
response = requests.get(url_metadata)
data = response.text
for i, line in enumerate(data.split('\n')):
    print(f'{i}   {line}')
```

```
df = pd.read_csv(url_data, header=None)
df.columns = [
  "Sex", "Length", "Diameter", "Height", "Whole weight", "Shucked weight", "Viscera weight", "Shell weight", "Rings"
]

df.info()
```
